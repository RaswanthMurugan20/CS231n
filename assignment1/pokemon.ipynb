```python
import numpy as np

def a():
    x,y = input("enter 2 numbers").split()
    floor = np.zeros((int(x),int(y)))
    return floor


def agent2():

    s = np.random.rand(1,1)
    if s>0.5:
        
        a1 = np.random.randint(0,1,size = 1)
        a2 = np.random.randint(-1,2,size = 1)
    else:
        
        a2 = np.random.randint(0,1,size = 1)

        a1 = np.random.randint(-1,2,size = 1)
    
    return [a1,a2];



def action():
    
    re = np.zeros((100,1))
    floo = a()
    
    q = np.random.randint(floo.shape[0])
    r = np.random.randint(floo.shape[1])
    
    for i in range(100):
      
        x = np.random.randint(floo.shape[0])
        y = np.random.randint(floo.shape[1])
        floo[x][y] += 1
        k = np.squeeze(np.asarray(agent2()),axis = (1,))
        if q + k[0] > floo.shape[0]-1 or q + k[0]< 0 or r + k[1] > floo.shape[1]-1 or r + k[1]< 0:
            re[i] = -10
        else:
            if k[0] == 0 and k[1] == 0:
                               
                if floo[q][r] == 0:
                    re[i] = -1
                    
                else:
                    
                    re[i] = floo[q][r]
                    floo[q][r] = 0
            q = q + k[0]
            r = r + k[1]
            
        print("({},{}) ({},{}) {}".format(k[0],k[1],q,r,re[i]))
      
    print(np.sum(re)/100) 
    
        
action()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
    

def ob(p_kid):
    if p_kid*100 > (math.ceil(p_kid*100)+math.floor(p_kid*100))/2:
        p_round = math.ceil(p_kid*100)
    else:
        p_round = math.floor(p_kid*100)
    
    x = np.random.randint(100,size = 1)
    
    if x>p_round:
        flag = 1
    else:
        flag = 0
    
    y = np.random.rand(1,1)
    
    if flag:
        
       
        
        if y < 0.5:
            
            obs_h = np.random.uniform(5,5.25,1)
            
        else:
                    
            obs_h = np.random.uniform(5.25,5.75,1)
            
        obs_w = np.random.uniform(45,60,1)
        
    else:
        
        if y < 0.5:
            
            obs_h = np.random.uniform(4.5,5,1)
            
        else:
                    
            obs_h = np.random.uniform(5,5.25,1)
        
        obs_w = np.random.uniform(30,50)
    
    return [obs_h,obs_w,flag];



def agent(h,w):
    
    if h<5 or w<45:
        a = 0
    elif h>5.25 or w>50:
        a = 1
    else:
        x = np.random.uniform(0,1,1)
        if x > 0.5:
            a  = 1
        else:
            a = 0
    return a


k = ob(0.5)
print("{}{}".format(k[0],k[1]))
print(agent(k[0],k[1]))


data = np.zeros((1000,4))
a = np.zeros((1000,1))
for i in range(1000):
    
    k = ob(0.5)
    data[i] = ob(0.5)+[i]
    a[i] = agent(k[0],k[1])
    

y = (data[:,2][:,np.newaxis] == a)

print(np.sum(y.astype(int))/1000)

plt.hist(data[:,0])
plt.show()


def envi():
    x,y = input("enter the number of rows and columns").split()
    
    e = np.zeros((int(x),int(y)))
    
    return e


floor = envi()
    
xt = np.random.randint(floor.shape[0])
yt = np.random.randint(floor.shape[1])

                    
def agent1(xt,yt):
    
    j = np.random.randint(-1,2,size = 2)
    
    return [j[0],j[1]];


reward = np.zeros((100,1))

print("{}{}".format(xt,yt));

for i in range(100):
    
    x = np.random.randint(floor.shape[0])
    y = np.random.randint(floor.shape[1])
    
    floor[x][y] += 1
    k = agent1(xt,yt)
    if k[0] == 0 and k[1] == 0:
        if floor[xt][yt] == 0:
            reward[i] = -1
        else:
            reward[i] = floor[xt][yt]
            floor[xt][yt] = 0
    else:
        if xt+k[0]>floor.shape[0]-1 or xt+k[0]<0 or yt+k[1]>floor.shape[1]-1 or yt+k[1]<0:
            reward[i] = -10
        else:
            xt = xt+k[0]
            yt = yt+k[1]
    
    print("({},{}) ({},{}) ({},{}) {}".format(xt,yt,x,y,k[0],k[1],reward[i]))
    print("\n")
    

print(np.sum(reward))
print(floor)
    

```
